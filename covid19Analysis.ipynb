```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df1 = pd.read_csv('confirmed.csv')
df2 = pd.read_csv('deaths.csv')
df3 = pd.read_csv('recovered.csv')
```

```
df_N_Confirmed = df1[df1['Country/Region']=='Nigeria']
df_N_Confirmed.drop(df_N_Confirmed.columns[[0,1,2,3]],axis=1,inplace=True)
df_N_recovered = df3[df3['Country/Region']== 'Nigeria']
df_N_recovered.drop(df_N_recovered.columns[[0,1,2,3]],axis=1,inplace=True)
df_N_deaths = df2[df2['Country/Region']== 'Nigeria']
df_N_deaths.drop(df_N_deaths.columns[[0,1,2,3]],axis=1,inplace=True)
```

```

confirmed = df_N_Confirmed.iloc[0]
deaths = df_N_deaths.iloc[0]
recovered = df_N_recovered.iloc[0]
```

```
plt.figure(figsize=(16,8)) 
confirmed.plot(kind='line', c='yellow', label='confirmed')
deaths.plot(kind="line", c='red',label='deaths')
recovered.plot(kind="line", c='green', label='recovered')
plt.grid(b=None,which='major', axis='both')

plt.xlabel('Date')
plt.ylabel('Confirmed cases')
plt.title('Nigeria covid confirmed cases')
plt.legend()
plt.savefig('covid2.png')

plt.show()
```
